```python
# Função algébrica da distância entre dois vetores

def distancia_vetores(a,b):
  return np.linalg.norm(a - b)
```

```python
# Programando a distância dos usuários com a condição de distancia minima para aparecer como 100000

def distancia_usuarios(user1, user2, minimo = 5):
  nota_1 = notas_do_usuario(user1)
  nota_2 = notas_do_usuario(user2)
  distancia_notas = nota_1.join(nota_2, lsuffix= "_esquerda", rsuffix= "_direita").dropna()

  if(len(distancia_notas) < minimo):
    return[user1, user2, 100000]

  distancia = distancia_vetores(distancia_notas['Nota_esquerda'], distancia_notas['Nota_direita'])
  return [user1, user2, distancia]
```

# Filtrando N usuários mais próximos

```python
# Retornando None (ao invés de 100000) para usuários com quantidade de filmes em comum menor que o mínimo

def distancia_usuarios(user1, user2, minimo = 5):
  nota_1 = notas_do_usuario(user1)
  nota_2 = notas_do_usuario(user2)
  distancia_notas = nota_1.join(nota_2, lsuffix= "_esquerda", rsuffix= "_direita").dropna()

  if(len(distancia_notas) < minimo):
    return None

  distancia = distancia_vetores(distancia_notas['Nota_esquerda'], distancia_notas['Nota_direita'])
  return [user1, user2, distancia]
```

```python
# Acrescentando filtro para retirar da tabela os valores None

def distancia_todos_usuarios(voce_id, numero_de_usuarios_a_analisar = None):
  todos_os_usuarios = notas['UsuarioId'].unique()
  if numero_de_usuarios_a_analisar:
    todos_os_usuarios = todos_os_usuarios[:numero_de_usuarios_a_analisar]
  distancia = [distancia_usuarios(voce_id, Usuario_id) for Usuario_id in todos_os_usuarios]
  distancia = list(filter(None, distancia))
  distancia = pd.DataFrame(distancia, columns = ["voce", "outra_pessoa", "distancia"])
  return distancia
```

```python
# Definindo usuários mais próximos com amostra do df

def mais_proximos_de(voce_id, numero_de_usuarios_a_analisar = None):
  distancia = distancia_todos_usuarios(voce_id, numero_de_usuarios_a_analisar = numero_de_usuarios_a_analisar)
  distancia = distancia.sort_values('distancia')
  distancia = distancia.set_index('outra_pessoa').drop(columns = "voce")
  distancia = distancia.drop(index=voce_id)
  return distancia
```

# Montando as recomendações

```python
def mais_proximos_de(voce_id, numero_de_usuarios_a_analisar = None):
  distancia = distancia_todos_usuarios(voce_id, numero_de_usuarios_a_analisar = numero_de_usuarios_a_analisar)
  distancia = distancia.sort_values('distancia')
  distancia = distancia.set_index('outra_pessoa').drop(columns = "voce")
  distancia = distancia.drop(index=voce_id)
  return distancia
```

```python
# Definindo parâmetro de recomendação pelo PRIMEIRO usuário mais próximo

def sugestoes_para(voce_id, numero_de_usuarios_a_analisar = None):
  notas_voce = notas_do_usuario(voce_id)
  filmes_assistidos = notas_voce.index

  similares = mais_proximos_de(voce_id, numero_de_usuarios_a_analisar = numero_de_usuarios_a_analisar)
  similar = similares.iloc[0].name #pegando o usuario mais similar pelo UsuarioId
  notas_do_similar = notas_do_usuario(similar).drop(filmes_assistidos, errors = 'ignore')
  recomendacoes = notas_do_similar.sort_values('Nota', ascending = False)

  return recomendacoes.join(filmes)
```

# Sugerindo com base em múltiplos usuários

```python
# Definindo quantos usuários mais próximos serão selecionados

def mais_proximos_de(voce_id, n_mais_proximos = 10, numero_de_usuarios_a_analisar = None):
  distancia = distancia_todos_usuarios(voce_id, numero_de_usuarios_a_analisar = numero_de_usuarios_a_analisar)
  distancia = distancia.sort_values('distancia')
  distancia = distancia.set_index('outra_pessoa').drop(columns = 'voce')
  distancia = distancia.drop(index=voce_id)
  return distancia.head(n_mais_proximos)
```

```python
def media_notas_mais_proximos(voce_id, n_mais_proximos = 10, numero_de_usuarios_a_analisar = None):
  similares = mais_proximos_de(voce_id, n_mais_proximos = n_mais_proximos, numero_de_usuarios_a_analisar = numero_de_usuarios_a_analisar)
  usuarios_similares = similares.index
  notas_dos_similares = notas.set_index("UsuarioId").loc[usuarios_similares]
  recomendacoes = notas_dos_similares.groupby("FilmeId").mean()[["Nota"]] # Feito a média das notas dadas pelos usuários mais próximos
  recomendacoes = recomendacoes.sort_values("Nota", ascending=False)
  return recomendacoes.join(filmes)
```

# Filtro mínimo de filmes em comum assistidos pelos usuários mais próximos

```python
# Recomendação final

def sugere_para(voce_id, n_mais_proximos = 10, numero_de_usuarios_a_analisar = None): #Selecionamos a amostra
  notas_de_voce = notas_do_usuario(voce_id)
  filmes_que_voce_ja_viu = notas_de_voce.index # Listamos os filmes já assistidos

  similares = mais_proximos_de(voce_id, n_mais_proximos = n_mais_proximos, numero_de_usuarios_a_analisar = numero_de_usuarios_a_analisar)
  usuarios_similares = similares.index # Listamos usuários similares
  notas_dos_similares = notas_com_mais_50_votos.set_index("UsuarioId").loc[usuarios_similares]
  recomendacoes = notas_dos_similares.groupby("FilmeId").mean()[["Nota"]]
  views = notas_dos_similares.groupby("FilmeId").count()[['Nota']]
  
  filtro_minimo = n_mais_proximos / 2 # Pelo menos 50% dos usuários devem ter assistido aos filmes recomendados
  recomendacoes = recomendacoes.join(views, lsuffix="_media_dos_usuarios", rsuffix="_quantidade_de_views")
  recomendacoes = recomendacoes.query("Nota_quantidade_de_views >= %.2f" % filtro_minimo) # Filmes vistos por mais de 50% dos 'mais próximos'
  recomendacoes = recomendacoes.sort_values("Nota_quantidade_de_views", ascending=False)
  recomendacoes = recomendacoes.drop(filmes_que_voce_ja_viu,errors='ignore') # Retiramos os filmes já assistidos
  recomendacoes = recomendacoes.join(filmes)
  recomendacoes = recomendacoes.drop(columns = ['Nota_media','Nota_quantidade_de_views','Total_votos'])
  return recomendacoes
```

```python
def mais_proximos_de(voce_id, n_mais_proximos=10, numero_de_usuarios_a_analisar = None):
  distancia = distancia_todos_usuarios(voce_id, numero_de_usuarios_a_analisar = numero_de_usuarios_a_analisar)
  distancia = distancia.sort_values("distancia")
  distancia = distancia.set_index("outra_pessoa").drop(voce_id, errors='ignore')
  distancia = distancia.drop(columns = 'voce')
  return distancia.head(n_mais_proximos)
```
